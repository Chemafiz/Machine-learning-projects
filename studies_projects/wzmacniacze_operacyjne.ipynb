```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
U_1 = np.concatenate(((np.arange(1,16), np.arange(-1, -15, -1)))) 
U_2 = np.array([0.945,
1.95,
2.95,
3.95,
4.96,
5.96,
6.96,
7.96,
8.97,
9.97,
10.97,
11.98,
12.98,
13.91,
13.99,               
-1.05,
-2.05,
-3.06,
-4.06,
-5.06,
-6.06,
-7.06,
-8.07,
-9.07,
-10.07,
-11.07,
-12.07,
-12.38,
-12.38,
])
```

```python
def fitowanie_krzywej(indeksy, wartosci):
    dane_z_biasem = np.c_[indeksy, np.ones((indeksy.size, 1))]
    parametry_krzywej= np.linalg.inv(dane_z_biasem.T @ dane_z_biasem) @ dane_z_biasem.T @ wartosci
    print(f"współczynnik nachylenia krzywej: {parametry_krzywej[0]} \nwspółczynnik bias: {parametry_krzywej[1]}")
    return parametry_krzywej
```

```python
parametry_krzywej = fitowanie_krzywej(U_1, U_2)
```

```python
X_new = np.array([-12 , 13])
X_new_b = np.concatenate((X_new.reshape(-1,1), np.ones((2,1))), axis=1)
y_predict= X_new_b @ parametry_krzywej
```

```python
plt.figure(figsize=(12,6))
plt.title("Wykres zależności napięcia wejściowego od wyjściowego", fontsize=20)
plt.scatter(U_1, U_2, s=40, c="k")
plt.plot(X_new, y_predict, "r", linewidth=2)
plt.plot([-15, 15], [0,0], "b--" )
plt.axis([-16, 16, -15, 16])
plt.xlabel("U_1[V]", fontsize=20)
plt.ylabel("U_2[V]", fontsize=20)
plt.show()
```
